```python
import pandas as pd

df = pd.read_csv('../data/fake_job_postings.csv')
print("Shape:", df.shape)
print("\nColumn names:")
print(df.columns.tolist())
```

```python
print("Fake vs Real count:")
print(df['fraudulent'].value_counts())

print("\nIn percentage:")
print((df['fraudulent'].value_counts(normalize=True) * 100).round(2))
```

```python
print("Missing values per column:")
print(df.isnull().sum())
```

```python
fake = df[df['fraudulent'] == 1][['title','description','requirements']].head(5)

for i, row in fake.iterrows():
    print("=" * 60)
    print("TITLE:", row['title'])
    print("\nDESC:", str(row['description'])[:400])
    print("\nREQ:", str(row['requirements'])[:200])
    print()
```

```python
real = df[df['fraudulent'] == 0][['title','description']].head(3)

for i, row in real.iterrows():
    print("=" * 60)
    print("TITLE:", row['title'])
    print("\nDESC:", str(row['description'])[:400])
    print()
```
